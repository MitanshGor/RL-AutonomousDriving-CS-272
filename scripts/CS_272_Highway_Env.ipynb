```python
%%writefile init_env_.py
import highway_env
import gymnasium as gym

def createEnv(ind, seed=42):
  def init_env():
    env = gym.make('highway-v0',
                    render_mode='rgb_array',
                    config = {
                          "observation": {
                              "type": "LidarObservation",
                              "cells": 128,
                              "maximum_range": 64,
                              "normalise": True,
                          },
                    })
    env.reset(seed=seed+ind)
    return env
  return init_env
```

```python
import sys
sys.path.append('/content/drive/MyDrive/PPO_Model')
import gymnasium as gym
from gymnasium.wrappers import RecordVideo
from stable_baselines3.common.env_util import make_vec_env
import json
import highway_env
from ppo import PPO
import os
import matplotlib.pyplot as plt
from stable_baselines3.common.results_plotter import load_results, ts2xy
import numpy as np
from stable_baselines3.common.evaluation import evaluate_policy
import pandas as pd
from stable_baselines3.common.callbacks import EvalCallback
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import SubprocVecEnv, VecMonitor
import multiprocessing as mp
from init_env_ import createEnv

if __name__ == "__main__":

    log_dir = '/content/drive/MyDrive/CS 272 Project/highway_lidar_results/'
    os.makedirs(log_dir, exist_ok=True)

    env = SubprocVecEnv([createEnv(i) for i in range(4)], start_method='fork')
    env  = VecMonitor(env, log_dir)

    model = PPO('CnnPolicy',
                env,
                policy_kwargs=dict(net_arch=[256,256,256]),
                batch_size=64,
                n_epochs=20,
                gamma=0.99,
                ent_coef=0.01,
                learning_rate=3e-4,
                kl_coef=1e-1,
                dual_clip=2,
                verbose=1,
                tensorboard_log=log_dir+'tensorboard/',
                device="cuda")

    eval_callback = EvalCallback(
            env,
            best_model_save_path=log_dir+'best_model/',
            log_path=log_dir+'ppo_eval/',
            eval_freq=10000
        )

    print("Environment created")

# Train the model
model.learn(total_timesteps=int(150_000), callback=eval_callback, progress_bar=True)
model.save(log_dir+"model/")

x, y = ts2xy(load_results(log_dir), 'timesteps')
y = np.convolve(y, np.ones(1000)/1000, mode='valid')
x = x[:len(y)]
plt.plot(x,y)
plt.xlabel('Timesteps')
plt.ylabel('Reward')
plt.title('PPO Highway Lidar Env Learning Curve')
plt.savefig(log_dir+'/ID1_highway_lidar_learning_curve.png')
plt.show()

rewards, lengths = evaluate_policy(model, env, n_eval_episodes=500, return_episode_rewards=True)
plt.violinplot(rewards, showmeans=True)
plt.ylabel('Reward')
plt.title('PPO Highway Lidar Env Performance 500 Episodes')
plt.savefig(log_dir+'/ID2_highway_lidar_performance_test.png')
plt.show()

env.close()
```
